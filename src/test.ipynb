```python
from matplotlib import pyplot as plt
import numpy as np 
from numpy import random as npr
```

```python
def choose_action(d, a, f):
    total = d + a + f
    d, a, f = d/total, a/total, f/total
    u = npr.rand()
    l = [d, d+a, 1.0]
    x = 0
    while u > l[x]:
        x += 1
    return x

size = 1000
d, a, f = 0.2, 0.5, 0.7

L = [choose_action(d,a,f) for _ in range(size)]

plt.hist(L, bins=[-0.5, 0.5, 1.5, 2.5], rwidth=0.8)
plt.xticks([0,1,2], ['diffusion','absorption','fission'])
plt.title("Distribution des actions choisies")
plt.show()
```
